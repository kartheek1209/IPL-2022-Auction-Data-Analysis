```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
ipl_data = pd.read_csv("ipl_2022_dataset.csv")
ipl_data.head(10)
ipl_data.Player.str.lower()
```

```python
ipl_data["Player"] = ipl_data.Player.str.lower()
```

```python
ipl_data
```

```python
#check how many rows and columns are there
ipl_data.shape
```

```python
#check how many rows are there per column and identify different types of data
ipl_data.info()
```

```python
#get all the column names
ipl_data.columns
```

```python
#remove the unnamed column as it is unnecesary
ipl_data.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
ipl_data.head(10)
```

```python
#to check the number of null values present in each column
ipl_data.isnull().sum()
```

```python
#to check which rows in column 'Cost IN $ (000)' are null
ipl_data[ipl_data["Cost IN $ (000)"].isnull()]
```

```python
#let us fill the null values in cost columns with zeros
ipl_data["COST IN ₹ (CR.)"] = ipl_data["COST IN ₹ (CR.)"].fillna(0)
ipl_data["Cost IN $ (000)"] = ipl_data["Cost IN $ (000)"].fillna(0)
#check output now
ipl_data.isnull().sum()
```

```python
#get details of players that didnt play in 2021 (check for null in 2021)
ipl_data[ipl_data["2021 Squad"].isnull()]
```

```python
ipl_data["2021 Squad"] = ipl_data["2021 Squad"].fillna("Not Participated")
#check output now
ipl_data.isnull().sum()
```

```python
#let us check the rows for which we have cost in rs greater than zero
ipl_data[ipl_data["COST IN ₹ (CR.)"]>0]
```

```python
#select unique team names corresponding to rows where cost >0
teams = ipl_data[ipl_data["COST IN ₹ (CR.)"]>0]["Team"].unique()
teams
```

```python
#let us add a 'status' column to our data set. If a player belongs to a team (i.e., if team is not null, replace with "sold")
#In the original data set, for rows with cost = null, the value under team column was "unsold"
#so we took the rows with names in team column and updated corresponding status column with sold for values matching with the names in 'teams' list
#for values in team column that doesnt match with values in 'teams' list (i.e., "unsold"), we simply copied the same into status column
ipl_data["status"] = ipl_data["Team"].replace(teams,"sold")
ipl_data.head(10)
```

```python
ipl_data
```

```python
#we are removing all the duplicate player names (if same name occurs twice, we remove both)
ipl_data[ipl_data["Player"].duplicated(keep=False)]
```

```python
ipl_data.shape
```

```python
#How many players have participated in 2022 auction
ipl_data.shape[0]
```

```python
#How many types of players have participated in ipl 2022 auction
types = ipl_data["TYPE"].value_counts()
types.reset_index()
```

```python
#we are plotting a pie chart
#first parameter is what you want to plot
#second parameter is what is the label or name for each pie
#labeldistance is how far should the name be from the graph
#autopct is the number of decimals in the percentage share
#startangle is just the orientation of the graph, 60 is a good value
plt.pie(types.values, labels=types.index, labeldistance=1.15, autopct='%1.2f%%', shadow=True, startangle=60)
plt.title("Role of player participated", fontsize = 15)
plt.plot()
```

```python
#let us see how many players were sold and unsold using a bar graph
#figsize gives the breadth and length of the graph
#countplot is used to draw graphs showing us different counts. In this case, it counts the different values under "Status" column
#plt.xlabel() is name of x axis
#plt.ylabel() is the name of y axis
#plt.title() is the title of the figure
#plt.plot() gives us the graph
plt.figure(figsize=(10,5))
fig = sns.countplot(ipl_data["status"],palette = ["Orange","Blue"])
plt.xlabel("Sold or Unsold")
plt.ylabel("Number of players")
plt.title("Sold vs Unsold", fontsize = 15)
plt.plot()

ipl_data.groupby("status")["Player"].count()

for p in fig.patches:
    fig.annotate(format(p.get_height(), '.0f'),(p.get_x() +
    p.get_width()/2.,p.get_height()), ha = 'center',va = 'center',xytext = (0,4),
    textcoords = 'offset points')
```

```python
#total number of players bought by each team
#plt.xticks() is used to rotate the x axis label for ease of understanding
plt.figure(figsize=(10,5))
fig = sns.countplot(ipl_data[ipl_data["Team"]!="Unsold"]["Team"])
plt.xlabel("Team Names")
plt.ylabel("Number of players")
plt.title("Players Bought By Each Team", fontsize = 15)
plt.xticks(rotation=70)
plt.plot()

ipl_data.groupby("Team")["Player"].count()

for p in fig.patches:
    fig.annotate(format(p.get_height(), '.0f'),(p.get_x() +
    p.get_width()/2.,p.get_height()), ha = 'center',va = 'center',xytext = (0,5),
    textcoords = 'offset points')
```

```python
ipl_data["retention"] = ipl_data["Base Price"]
```

```python
ipl_data
```

```python
#get unique values of retention column
ipl_data["retention"].unique()
```

```python
#replace the money with "From Auction"
ipl_data["retention"].replace(['2 Cr', '40 Lakh', '20 Lakh', '1 Cr', '75 Lakh',
       '50 Lakh', '30 Lakh', '1.5 Cr'],"From Auction",inplace = True)
ipl_data["retention"].unique()
```

```python
#replace "Draft PicK" in base price column with zero
ipl_data["Base Price"].replace("Draft Pick",0, inplace = True)
ipl_data["Base Price"].unique()
```

```python
#now we will split the base price column into 'base_price' and 'base_price_unit'
#base_price will have numerical value and base_price_unit will have lakh, cr etc
ipl_data["base_price_unit"] = ipl_data["Base Price"].apply(lambda x: str(x).split(' ')[-1])
ipl_data["base_price"] = ipl_data["Base Price"].apply(lambda x: str(x).split(' ')[0])
ipl_data
```

```python
#replace "Retained" in base_price with zero
ipl_data["base_price"].replace("Retained",0,inplace = True)
ipl_data["base_price"].unique()
```

```python
#total players who were retained and bought
#using [:-1] so that unsold players are not seen in the result set
ipl_data.groupby(["Team","retention"])["retention"].count()[:-1]
```

```python
#same bar graph as above, but with different types of players per team
plt.figure(figsize=(20,10))
fig = sns.countplot(ipl_data[ipl_data["Team"]!="Unsold"]["Team"], hue = ipl_data["TYPE"])
plt.xlabel("Team Names")
plt.ylabel("Number of players")
plt.title("Players Bought By Each Team", fontsize = 15)
plt.xticks(rotation=70)
plt.plot()

ipl_data.groupby("Team")["Player"].count()

for p in fig.patches:
    fig.annotate(format(p.get_height(), '.0f'),(p.get_x() +
    p.get_width()/2.,p.get_height()), ha = 'center',va = 'center',xytext = (0,5),
    textcoords = 'offset points')
```

```python
#Highest amount spent on a player by each team
#consider only the rows which have "From Auction" (we dont want retained players)
#groupby the resultset by "Team" and then find out the max value of cost in cr for that team and sort in descending order
ipl_data[ipl_data["retention"]=="From Auction"].groupby(["Team"])["COST IN ₹ (CR.)"].max()[:-1].sort_values(ascending = False)
```

```python
#now lets see players retained at max price by teams
ipl_data[ipl_data["retention"]=="Retained"].groupby(["Team"])["COST IN ₹ (CR.)"].max()[:-1].sort_values(ascending = False)
```

```python
#top 3 players retained for max price
ipl_data[ipl_data["retention"]=="Retained"].sort_values(by = "COST IN ₹ (CR.)", ascending = False).head(3)
```

```python
#top 5 batters bought in this auction
ipl_data[(ipl_data["retention"]=="From Auction") & (ipl_data["TYPE"]=="BATTER")].sort_values(by = "COST IN ₹ (CR.)", ascending = False).head(5)
```

```python
#top 5 ALL-ROUNDERS bought in this auction
ipl_data[(ipl_data["retention"]=="From Auction") & (ipl_data["TYPE"]=="ALL-ROUNDER")].sort_values(by = "COST IN ₹ (CR.)", ascending = False).head(5)
```

```python
#top 5 wicket keepers bought in this auction
ipl_data[(ipl_data["retention"]=="From Auction") & (ipl_data["TYPE"]=="WICKETKEEPER")].sort_values(by = "COST IN ₹ (CR.)", ascending = False).head(5)
```
